```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = [1, 2, 3, 4, 5]
y = [0.58, 0.63, 1.20, 2.54, 0.83]

# plot
fig, ax = plt.subplots()

plt.figure(figsize=(300,10))
ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 5), xticks=range(6),
       ylim=(0, 3), yticks=np.arange(0, 3, 0.1))


plt.show()
```
